#4B


```python
import nltk
from nltk.corpus import brown

# Make sure the Brown corpus is downloaded
nltk.download('brown')

# Build a Conditional Frequency Distribution:
# condition = genre, event = word
cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)

# List all conditions (genres)
cfd.conditions()

# Tabulate frequencies of selected words in selected genres
cfd.tabulate(
    conditions=['government', 'humor', 'reviews'],
    samples=['leadership', 'worship', 'hardship']
)

# Plot cumulative frequencies
cfd.plot(
    conditions=['government', 'humor', 'reviews'],
    samples=['leadership', 'worship', 'hardship'],
    cumulative=True
)

# Access the frequency distribution for the 'news' genre
news_fd = cfd['news']

# Most common 3 words in 'news'
news_fd.most_common(3)

# Frequency of the word 'the' in 'news'
news_fd['the']
```

#4C

```python
import nltk
from nltk.corpus import names

# Make sure the corpus is available
nltk.download('names')

# Build (gender, last_letter) pairs
nt = [
    (fid.split('.')[0], name[-1])
    for fid in names.fileids()
    for name in names.words(fid)
]

# Create Conditional Frequency Distribution
cfd2 = nltk.ConditionalFreqDist(nt)

# Frequency of names ending with 'a'
cfd2['female']['a']
cfd2['male']['a']

# Tabulate frequencies for selected last letters
cfd2.tabulate(samples=['a', 'e'])

# Plot last-letter distribution
cfd2.plot()

```

```python
def cal_average(num):
  sum_num = 0
  for t in num:
    sum_num = sum_num + t
    avg = sum_num / len(num)
  return avg
print("The average is", cal_average([18,25,3,41,5]))
```
